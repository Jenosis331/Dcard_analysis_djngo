```python
import pandas as pd

# 讀取（一定要加 sep="|"）
df = pd.read_csv("cna_news_preprocessed_all.csv", sep="|")

# 刪除 category 為 "暫無" 的列
df = df[df["category"] != "暫無"]
df = df[df["category"] != "('淡江'"] 
df["link"] = df["link"].astype(str).str.strip()

df = df[df["link"].str.startswith("/news/")]
# 存回檔案
df.to_csv("cna_news_preprocessed_all.csv", sep="|", index=False)
```

```python
import pandas as pd

df = pd.read_csv("cna_news_preprocessed_all.csv", sep="|")

# 清理欄位
df.columns = df.columns.str.strip()

df["category"] = df["category"].astype(str).str.strip()

df["link"] = (
    df["link"]
    .astype(str)
    .str.strip()
    .str.replace("\xa0", "", regex=False)
    .str.replace("　", "", regex=False)
)

# 篩選
df = df[df["category"] != "暫無"]
df = df[df["category"] != "('淡江'"]

# 👉 核心：只留正確 link
df = df[df["link"].str.match(r"^/news/[a-z]+/\d+\.aspx$")]

# 存檔
df.to_csv("cna_news_preprocessed_all.csv", sep="|", index=False)
```
